```python
df2 = df.iloc[:, 1:3]
df2 = df2.rename(columns={'英文':'en','簡中':'zh'})
df2 = df2.reindex(columns=['zh','en'])
df2 = df2[df2.zh.notna()]
df2.head()
```

```python
df = pd.concat([df1, df2], axis=0).reset_index(drop=True)
df.shape
```

```python
df.to_json('/Users/alexlo/Desktop/Project/Chai_Trans/data/workdata/term_format_zhen.json', orient='records', lines=True)
```
