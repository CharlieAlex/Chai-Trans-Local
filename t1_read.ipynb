```python
# 設定目標
directory = 'TMX'
file_name = 'ZH-EN (Charlotte).tmx'
from_lang = 'zh'
to_lang = 'en'
```

```python
# package
from library.read import ReadMachine

params = {
    'directory': directory,
    'file_name': file_name,
    'from_lang': from_lang,
    'to_lang': to_lang
}

read_machine = ReadMachine(params)
```

```python
# tmx to json
df = read_machine.tmx2df()
read_machine.df2json(df)
```

```python
# xlsx to json
df = read_machine.xlsx2df()
df = read_machine.repeat_df(df, 5)
read_machine.df2json(df)
```

```python
# csv to json
df = read_machine.csv2df()
read_machine.df2json(df)
```
